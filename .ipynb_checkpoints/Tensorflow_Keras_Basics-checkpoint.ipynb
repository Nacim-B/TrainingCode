```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Importing BTC and ETH data from Yahoo Finance. Datas are from April 2022 to March 2023
bitcoin = pd.read_csv('data/BTC-USD.csv', index_col='Date', parse_dates=True)
ethereum = pd.read_csv('data/ETH-USD.csv', index_col='Date', parse_dates=True)

#Creating a unique dataframe with ETH and BTC values, DATES set as the index
btc_eth = pd.merge(bitcoin, ethereum, on='Date', how='inner', suffixes=('_btc', '_eth'))
```

```python
#Correlation bitcoin BTC and ETH since April 2022
btc_eth[['Close_btc', 'Close_eth']].corr()
```

```python
# 2 Plots about the closing values of BTC and ETH since April 2022
btc_eth[['Close_btc','Close_eth']].plot(subplots=True)
```

```python
#Bitcoin prices with its every-28-day minimums (blue) and maximums (yellow)
bitcoin['Close'].shift(1).rolling(window=28).min().plot()
bitcoin['Close'].shift(1).rolling(window=28).max().plot()
bitcoin['Close'].plot()
```

```python
#Creates a copy of the original bitcoin dataframe
trading_strategy = bitcoin.copy()

#Adds 2 colums Sell and Buy, initialized at 0
trading_strategy['Buy'] = np.zeros(len(bitcoin))
trading_strategy['Sell'] = np.zeros(len(bitcoin))

#Adds 2 columns with every 28-day minimums/maximums
trading_strategy['RollingMax'] = trading_strategy['Close'].shift(1).rolling(window=28).max()
trading_strategy['RollingMin'] = trading_strategy['Close'].shift(1).rolling(window=28).min()

#Sets the Buy value to 1 when BTC price crosses above the 28d maximum
#Sets the Sell value to -1 when BTC pric crosses under the 28d minimum
trading_strategy.loc[trading_strategy['Close'] > trading_strategy['RollingMax'], 'Buy'] = 1
trading_strategy.loc[trading_strategy['Close'] < trading_strategy['RollingMin'], 'Sell'] = -1
```

```python
#Graphic that shows when to buy or sell BTC, based on the Turtle Strategy
plt.figure()
trading_strategy['Sell'].plot(c='r')
trading_strategy['Buy'].plot(c='g')
```
